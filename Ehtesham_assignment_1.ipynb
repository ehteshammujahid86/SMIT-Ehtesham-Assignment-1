```python
# 1) Write a program to take user input and check whether the number is positive, if it's check whether it is single, double, or three digits

number = input("Input a number")

if int(number) >= 0:
    if len(number)==1:
        print("The number is positive and single digit")
    elif len(number)== 2:
        print("The number is positive and double digit")
    else:
        print("The number is positive and three or more digit")
else:
    print("The number is negative try with positive number")
```
